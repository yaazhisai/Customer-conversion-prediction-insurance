```python
def iqrmeth(df,col):
    q3= df[col].quantile(0.75)
    q1= df[col].quantile(0.25)
    iqr=q3-q1
    upperbound=q3+(1.5*iqr)
    lowerbound=q1-(1.5*iqr)
    df[col]=np.where(df[col]>upperbound,upperbound,np.where(df[col]<lowerbound,lowerbound,df[col]))  



col=['age_log','dur_log','num_calls_log']
for i in col:
     iqrmeth(df1,i)
```

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc , accuracy_score
from sklearn.neighbors import KNeighborsClassifier


```

```python
def classification(x,y,model):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
    # Apply oversampling using RandomOverSampler
    o_s = RandomOverSampler(sampling_strategy='auto', random_state=42)
    X_train_oversampled, y_train_oversampled = o_s.fit_resample(x_train, y_train)
    # Apply undersampling using RandomUnderSampler
    u_s = RandomUnderSampler(sampling_strategy='auto', random_state=42)
    X_train_undersampled, y_train_undersampled = u_s.fit_resample(x_train, y_train)
    
    model_org = model().fit(x_train, y_train)

    # Fit on the oversampled train set
    model_os=model().fit(X_train_oversampled, y_train_oversampled)

    # Fit  on the undersampled train set
    model_us=model().fit(X_train_undersampled, y_train_undersampled)

    # Make predictions on train sets
    y_train_pred_original =model_org.predict(X_train)
    y_train_pred_oversampled =model_os.predict(X_train_oversampled)
    y_train_pred_undersampled =model_us.predict(X_train_undersampled)

    # Make predictions on test sets
    y_test_pred_original = model_org.predict(X_test)
    y_test_pred_oversampled =model_os.predict(X_test)
    y_test_pred_undersampled =model_us.predict(X_test)

    # Calculate and print accuracy for train sets
    print("MODEL",model)
    print("Accuracy on Original Train Set:", metrics.accuracy_score(y_train, y_train_pred_original))
    print("Accuracy on Oversampled Train Set:", metrics.accuracy_score(y_train_oversampled, y_train_pred_oversampled))
    print("Accuracy on Undersampled Train Set:", metrics.accuracy_score(y_train_undersampled, y_train_pred_undersampled))
    
    # Calculate and print accuracy for test sets
    print("\nAccuracy on Original Test Set:", metrics.accuracy_score(y_test, y_test_pred_original))
    #print (classification_report(y_test, y_test_pred_original))
    print("Accuracy on Oversampled Test Set:",metrics.accuracy_score(y_test, y_test_pred_oversampled))
    #print (classification_report(y_test, y_test_pred_oversampled))
    print("Accuracy on Undersampled Test Set:",metrics.accuracy_score(y_test, y_test_pred_undersampled))
    #print (classification_report(y_test, y_test_pred_undersampled))

    # evolutiom metrics for original
    MAE_org= metrics.mean_absolute_error(y_test, y_test_pred_original)
    MSE_org= metrics.mean_squared_error(y_test,y_test_pred_original)
    RSE_org=np.sqrt(metrics.mean_squared_error(y_test,y_test_pred_original))
    MAE_os= metrics.mean_absolute_error(y_test, y_test_pred_oversampled)
    MSE_os= metrics.mean_squared_error(y_test,y_test_pred_oversampled)
    RSE_os=np.sqrt(metrics.mean_squared_error(y_test,y_test_pred_oversampled))
    MAE_us= metrics.mean_absolute_error(y_test, y_test_pred_undersampled)
    MSE_us= metrics.mean_squared_error(y_test,y_test_pred_undersampled)
    RSE_us=np.sqrt(metrics.mean_squared_error(y_test,y_test_pred_undersampled))
    print("ORIGINAL DATA ")
    print("MSE",MSE_org)
    print("MAE",MAE_org)
    print("RSE",RSE_org)
    print("OS DATA ")
    print("MSE",MSE_os)
    print("MAE",MAE_os)
    print("RSE",RSE_os)
    print("US DATA ")
    print("MSE",MSE_us)
    print("MAE",MAE_us)
    print("RSE",RSE_us)

    

    
    
```

Insights:
Impact of Sampling Techniques:
Oversampling: While oversampling aims to address class imbalance by replicating minority class samples, it might lead to overfitting on the oversampled data or not generalize well.
Undersampling: Undersampling aims to balance the dataset by reducing the majority class, but it might lead to loss of valuable data, potentially reducing performance.
Model Performance:
Fitting the model without resampling shows good performance of the model, because accuracy is misleading with imbalanced data.
In this case, the training accuracy is slightly higher than the test accuracy,.
but the difference is relatively small, so severe overfitting is not indicated.
